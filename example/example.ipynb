```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arxiv_notebook import notebook_to_arxiv
from traitlets.config import Config
```

# Generate example data

In this section, we generate a DataFrame of two sine waves using `pandas` and `matplotlib`.

```python
# Generate two sine waves.
x = np.linspace(0, 4 * np.pi, 100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y1': np.sin(x), 'y2': np.sin(2 * x)}).set_index(x)
df
```

# Plot data

In this section we plot the sine waves.

```python
fig, ax = plt.subplots(1, 1)
df['y1'].plot(ax=ax)
df['y2'].plot(ax=ax)
```
